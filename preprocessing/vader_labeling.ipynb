**Analyze Sentiment Function**

function ini akan mengembalikan label sentiment berdasarkan analisis menggunaka VADER, menggunakan SentimentIntensityAnalyzer untuk menghitung score sentiment, kemudian mengecek skor keseluruhan ('compound_score') dan mengembalikan label yang sesuai dengan threshold yang ditentukan

Keterangan Label :
*   0 : Negative
*   1 : Netral
*   2 : Positive






```python
# Object initialisation
sid = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(comment, positive_threshold=0.05, negative_threshold=-0.05):
  sentiment_score = sid.polarity_scores(comment)
  compound_score = sentiment_score['compound']

  if compound_score >= positive_threshold:
    return 2
  elif compound_score <= negative_threshold:
    return 0
  else:
    return 1;
```

**Proses Labaling**

Melakukan proses labeling menggunakan VADER

```python
# Create a new column to store the labeling result
data['Sentiment'] = ''

# Perform the labeling process
for index, row in data.iterrows():
  sentiment = analyze_sentiment(row['Comment'])
  data.at[index, 'Sentiment'] = sentiment
```
